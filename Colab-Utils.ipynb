```python
import requests
import ipywidgets as widgets
from google.colab import drive

drive.mount('/content/gdrive')

widgets.BoundedIntText(value=7,min=0,step=1,description='CivitAI Model ID:', disabled=False)
```

```python


model_id = "29460"  # replace with actual model ID
url = f"https://civitai.com/api/download/models/{model_id}"

save_path = "/content/gdrive/MyDrive/sd/stable-diffusion-webui/models/Stable-diffusion/model.safetensors"

response = requests.get(url, stream=True)
response.raise_for_status()

with open(save_path, "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        if chunk:
            f.write(chunk)

print("Download complete:", save_path)
```
